```python
!pip install langchain-community -q

from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory
from google.colab import userdata

# 1. API 키 및 모델 설정
os.environ["GROQ_API_KEY"] = userdata.get('groq')
model = ChatGroq(model="llama-3.1-8b-instant")

# 2. 대화 기록을 세션별로 저장할 딕셔너리와 관리 함수
store = {}

def get_session_history(session_id: str) -> ChatMessageHistory:
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]

# 3. 랭체인의 래퍼(Wrapper)를 씌워 메모리 자동화 엔진 장착
bot_with_history = RunnableWithMessageHistory(
    model,
    get_session_history,
)

# --- 🚀 실전 대화 테스트 (우리는 오직 질문만 던집니다) ---
config = {"configurable": {"session_id": "user_123"}}

# 첫 번째 질문
response1 = bot_with_history.invoke([HumanMessage(content="내 이름은 민수야. 서울에 살고 있어.")], config=config)
print("AI:", response1.content)

print("-" * 30)

# 두 번째 질문 (AI가 위에서 답변한 내용을 기억하고 동작함!)
response2 = bot_with_history.invoke([HumanMessage(content="내 이름이 뭐라고 했지? 어디 산다고 했어?")], config=config)
print("AI:", response2.content)
```

```python
# @title Getting Specific Formats out of LLMs
from pydantic import BaseModel

class AnswerWithJustification(BaseModel):
    """An answer to the user's question along with justification for the answer."""

    answer: str
    """The answer to the user's question"""
    justification: str
    """Justification for the answer"""

llm = ChatGroq(model="llama-3.1-8b-instant", temperature=0)

structured_llm = llm.with_structured_output(AnswerWithJustification)

response = structured_llm.invoke(
    "What weighs more, a pound of bricks or a pound of feathers")
print(response)
```

## **Assembling the Many Pieces of an LLM Application**
- Runnable = 하나의 작업(함수)이 객체화 된 것
> PromptTemplate, ChatGroq(ChatGoogleGenerativeAI, ChatOpenAI), OutputParser --> 모두 Runnable Intergace (method가 동일함, runnable.invoke() , runnable.batch(), runnable.stream())
- LCEL = Runnable들을 연결하는 파이프라인 문법
> chain = prompt | llm | parser --> Unix의 pipe와 동일


```python
# @title Using the Runnable Interface
# @markdown - invoke() takes a single input and returns a single output.
# @markdown - batch() takes a list of outputs and returns a list of outputs.
# @markdown - stream() takes a single input and returns an iterator of parts of the output.

completion = model.invoke("Hi there!")
# Hi!
print(completion)
```

```python
print(completion.content)
```

```python
completions = model.batch(["Hi there!", "Bye!"])
# To print the content of each message:
for completion in completions:
    print(completion.content)
```

```python
for token in model.stream("Bye!"):
    print(token)
```

### Runnable들을 연결하는 2가지 방법
1.   Imperative: invoke()를 직접 하나씩 호출하며 실행 순서를 프로그래머가 명시
2.   Declarative(LCEL): prompt | llm | parser처럼 작업 흐름만 선언하면 LangChain이 실행 순서를 관리하고 필요한 기능(병렬 실행, 스트리밍, 비동기 등)을 일관된 방식으로 처리

> 두 방식의 결과는 같지만, Imperative는 "실행 방법을 직접 코딩"하는 방식(사용자정의 함수 실행)이고, LCEL은 "데이터 흐름을 선언"하는 방식이라는 점이 가장 큰 차이

- **Runnable Interface가 AsyncIO를 지원(LangChain은 OpenAI, Gemini, Anthropic, Ollama, Pinecone, Chroma, PostgreSQL 등 수많은 컴포넌트를 직렬 혹은 병렬로 연결해야 하기에 네트워크 요청(IO)은 비동기 방식을 지원)**

```python
# @title Imperative Composition
# @markdown - Imperative composition is just a fancy name for writing the code you’re used to writing, composing these components into functions and classes. Here’s an example combining prompts, models, and output parsers:


from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import chain

# the building blocks

template = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a helpful assistant."),
        ("human", "{question}"),
    ]
)

# combine them in a function
# @chain decorator adds the same Runnable interface for any function you write
# @chain 덕분에 chatbot 함수는 단순한 함수가 아니라 .
# invoke() 메서드를 사용할 수 있는 스마트한 LangChain 컴포넌트 역할

@chain
def chatbot(values):
    prompt = template.invoke(values)
    return model.invoke(prompt)

# use it
response = chatbot.invoke({"question": "Which model providers offer LLMs?"})
print(response.content)
```

```python
# @title Declarative Composition
# @markdown LCEL is a declarative language for composing LangChain components. LangChain compiles LCEL compositions to an optimized execution plan, with automatic paralleli‐zation, streaming, tracing, and async support.
template = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a helpful assistant."),
        ("human", "{question}"),
    ]
)


@chain
def chatbot(values):
    prompt = template.invoke(values)
    for token in model.stream(prompt):
        yield token


for part in chatbot.stream({"question": "Which model providers offer LLMs?"}):
    print(part)
```

```python
@chain
async def chatbot(values):
    prompt = await template.ainvoke(values)
    return await model.ainvoke(prompt)


async def main():
    return await chatbot.ainvoke({"question": "Which model providers offer LLMs?"})


# In Colab, we can directly await top-level async functions.
print(await main())
```

```python
# combine them with the | operator
chatbot = template | model

# use it
response = chatbot.invoke({"question": "Which model providers offer LLMs?"})
print(response.content)
```

```python
# streaming (token 형태로 출력)
for part in chatbot.stream({"question": "Which model providers offer LLMs?"}):
    print(part)
```
